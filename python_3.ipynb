```python
hari = "selasa"
match hari:
  case "senin":
    print("upacara")
  case "selasa":
    print("senam")
  case "rabu":
    print("literasi")
  case "kamis":
    print("kamis ceria")
  case "jumat":
    print("imtaq")
  case "sabtu":
    print("libur")
  case "minggu":
    print("libur")
```

```python
barang = "B01"
match barang:
  case "A01":
    print("snack=10000")
  case "A02":
    print("minuman=8000")
  case "B01":
    print("mie instan=3000")
  case "B02":
    print("sabun=5000")
  case "C01":
    print("shampoo=12000")
```

```python
while True :
  kode = input("input")

  match kode :
      case"A01":
         print ("nama barang : snack")
         print ("harga : 10000")
      case"A02":
         print ("nama barang : minuman")
         print ("harga : 8000")
      case"B01":
         print ("nama barang : mie instan")
         print ("harga : 3000")
      case"B02":
         print ("nama barang : sabun")
         print ("harga : 5000")
      case"C01":
         print ("nama barang : shampoo")
         print ("harga : 12000")
```
